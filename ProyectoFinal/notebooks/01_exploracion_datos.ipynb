# 01 · Exploración de datos (EDA)

Objetivo: comprender el corpus Fake & Real News antes de modelar.

- Tamaño y balance de clases
- Longitud de textos
- Distribución por `subject` y por fecha
- Word clouds por clase (opcional)

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_loader import load_fake_real_news
from src import config

sns.set_theme(style='whitegrid')
df = load_fake_real_news()
df.shape, df['label'].value_counts().to_dict()
```

```python
df.head(3)
```

## Balance de clases

```python
fig, ax = plt.subplots(figsize=(5, 3))
df['label'].map({0:'FAKE', 1:'REAL'}).value_counts().plot.bar(ax=ax, color=['#d9534f','#5cb85c'])
ax.set_title('Distribución de clases')
ax.set_ylabel('Cantidad')
plt.tight_layout()
plt.savefig(config.FIGURES_DIR / 'eda_balance.png', dpi=120)
plt.show()
```

## Longitud de textos

```python
df['n_chars'] = df['text'].str.len()
df['n_words'] = df['text'].str.split().str.len()

fig, axes = plt.subplots(1, 2, figsize=(11, 3.5))
for ax, col, title in zip(axes, ['n_chars','n_words'], ['Caracteres','Palabras']):
    sns.histplot(data=df, x=col, hue='label', bins=60, ax=ax, palette=['#d9534f','#5cb85c'])
    ax.set_title(f'Distribución de {title.lower()}')
plt.tight_layout()
plt.savefig(config.FIGURES_DIR / 'eda_longitudes.png', dpi=120)
plt.show()
```

## Distribución por tema (`subject`)

```python
df.groupby(['subject','label']).size().unstack(fill_value=0).plot.barh(figsize=(7,4), stacked=True, color=['#d9534f','#5cb85c'])
plt.title('Subject vs. clase')
plt.tight_layout()
plt.savefig(config.FIGURES_DIR / 'eda_subject.png', dpi=120)
plt.show()
```
