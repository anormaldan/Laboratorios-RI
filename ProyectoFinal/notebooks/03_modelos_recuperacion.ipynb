# 03 · TF-IDF y BM25

Construcción de los dos recuperadores clásicos y consultas de ejemplo.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from src import config
from src.retrieval import TfidfRetriever, BM25Retriever

df = pd.read_parquet(config.CORPUS_PATH)
print('Corpus:', len(df))
```

```python
tfidf = TfidfRetriever().fit(df['clean_text'].tolist())
tfidf.save(config.TFIDF_INDEX_PATH)

bm25 = BM25Retriever().fit(df['clean_text'].tolist())
bm25.save(config.BM25_INDEX_PATH)
print('Índices guardados.')
```

```python
query = 'trump election fraud voting machines'
for retriever in (tfidf, bm25):
    idx, scores = retriever.search(query, top_k=5)
    print(f'\n=== {retriever.name.upper()} ===')
    for r, (i, s) in enumerate(zip(idx, scores), start=1):
        title = df.iloc[i]['title'][:90]
        label = 'REAL' if df.iloc[i]['label'] == 1 else 'FAKE'
        print(f' [{r}] score={s:.3f} {label}  {title}')
```
